# Planet Subscription (korak za korakom)

Ta zvezek pojasni, kako v uporabniškem vmesniku Planet Insights ustvarite Planet Subscription:
https://insights.planet.com/data/subscriptions

Ciljne nastavitve v tem vodiču:
- Tip posnetka: `PSScene`
- Tip produkta (asset): `ortho_analytic_8b_sr`
- Časovni razpon: od `2022-01-01` naprej
- Filtri/orodja:
  - oblačnost manj kot 20 %
  - obrezovanje na območje
  - harmonizacija na S2 po metodi Chastain et al.

## Predpogoji

Pred začetkom preverite, da imate:
- Planet račun z dovoljenji za ustvarjanje naročnin
- Datoteko AOI (GeoJSON) ali AOI, narisan na zemljevidu
- Izbran cilj dostave (na primer Sentinel Hub kolekcija)

Referenčna dokumentacija:
- UI: https://insights.planet.com/data/subscriptions
- API dokumentacija: https://docs.planet.com/develop/apis/subscriptions/

## Odprite Subscriptions in začnite novo naročnino

1. Odprite https://insights.planet.com/data/subscriptions
2. Kliknite **Create subscription**.
3. Vnesite jasno ime, na primer:
   - `Kras-2022-onward`
   - `Marezige-2023-onward`

![Začetna stran Subscriptions v Planet Insights.](slike/planet_sub_land.png)

Vpišite ime, ki jasno opisuje vašo naročnino. To bo olajšalo prepoznavanje naročnine v prihodnosti, še posebej, če boste imeli več naročnin.

![alt text](slike/planet_sub_new.jpg)

Izberite:
* `PSScene` in
* `ortho_analytic_8b_sr`.

Če vključite **Harmonize**, dodajte potrebujete tudi spremljevalne produkte `ortho_analytic_8b_xml` in `ortho_udm2`. saj orodje potrebuje metapodatke in podatke kakovosti.

### Določite AOI

V razdelku AOI območje narišite na zemljevidu ali naložite/izberite GeoJSON. Oziroma ga izberite iz knjižnice AOI, če ste ga že prej naložili.

![alt text](slike/planet_sub_aoi.jpg)

Pred nadaljevanjem vizualno preverite mejo AOI.

### Nastavite časovni razpon

Izberite časovni razpon, ki ustreza vašim potrebam, začetni datum nastavite na `2022-01-01` ali pozneje, odvisno od tega, kdaj želite začeti prejemati posnetke. Končni datum pustite prazen (`ongoing`), če želite prejemati tudi prihodnje posnetke. Končni datum nastavite le, če potrebujete naročnino samo za arhivske podatke.

![alt text](slike/planet_sub_filters.jpg)

### Dodajte filtre

Pri dodajanju filtrov upoštevajte, da strožji filtri zmanjšajo število posnetkov, ki jih boste prejeli, vendar povečajo kakovost. Priporočamo, da začnete z zmernimi nastavitvami in jih prilagodite glede na rezultate.

**Clip** vedno deluje na AOI, je privzeto vključen in ni mogoče nastaviti pogojev.

Dodajte filter za `cloud_cover`, nastavite pogoj na **manj ali enako**. Priporočena privzeta vrednost: `0.20` (20 %). Namig: za strožjo kakovost znižajte na `0.10`, za več zadetkov povečajte na `0.30`.

Vključite **Harmonize** na `Sentinel-2` (če potrebujete konsistentnost med senzorji).


## Nastavite dostavo in ustvarite naročnino

Preglejte povzetek:
   - PSScene
   - ortho_analytic_8b_sr (ter zahtevani dodatni asseti, če uporabljate harmonizacijo)
   - Začetni datum 2022-01-01
   - Oblačnost <= 0.20
   - Harmonize in Clip po potrebi

Kliknite **Subscribe**.

## Preverite naročnino

Po ustvarjanju:
1. Preverite, da se status premakne v `active/running`.
2. Potrdite, da podatki prihajajo v izbrani cilj dostave.
3. Na vzorčnih prizorih preverite učinek clip in harmonize.

![alt text](slike/planet_sub_final.png)

Spremljajte izvajanje.

![alt text](slike/planet_sub_running.png)

Prikažite in analizirajte prejeti posnetke, da potrdite, da naročnina deluje po pričakovanjih. Če opazite težave, se vrnite na nastavitve naročnine in jih prilagodite po potrebi.

![alt text](slike/planet_sub_analyse.jpg)
